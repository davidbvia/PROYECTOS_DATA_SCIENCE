```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from time import sleep
```

```python
url = "https://www.info-computer.com/portatiles"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
urls = [s["href"] for s in soup.find_all("a", href = True)]
```

```python
# Se extraen las urls que contengan portátiles y los datos de cada uno de ellos
urls, especificacion, precio = list(), list(), list()

for i in range(1, 45):
    urls = f"https://www.info-computer.com/portatiles/?page={i}"
    response = requests.get(urls)
    soup = BeautifulSoup(response.text, "html.parser")
    for urls in soup.find_all("div", class_ = "thumbnail-container"):
        try:
            especificaciones = urls.find("p", class_ = "h3 product-title").text
            precios = urls.find("span", class_ = "price").text
            especificacion.append(especificaciones)
            precio.append(precios)
        except:
            print("Sin especificaciones")    
```

```python
# Se preparan los datos antes de crear la lista de listas 
modeloProc, ram, disco, webcam, sistema = list(), list(), list(), list(), list()
lista_separada, lista_final = list(), list()
contador = 0
for i in especificacion:
    lista_separada = especificacion[contador].split(" | ")
    contador = contador + 1
    for j in lista_separada:
        modeloProc.append(lista_separada[:1])
        ram.append(lista_separada[1:2])
        disco.append(lista_separada[2:3])
        webcam.append(lista_separada[3:4])
        sistema.append(lista_separada[4:5])

# Se limpian los elementos de la columna precios
new_precio = list()
for p in precio:
    new_precio.append(precios.replace("\n", "").replace("\xa0€", "€"))

# Se terminan de agrupar los datos
portatiles_join = zip(modeloProc, ram, disco, webcam, sistema, webcam, new_precio)
portatiles = list(portatiles_join)
```

```python
# Pruebas
columns = ["Modelo y procesador", "RAM", "Disco y formato", "Webcam", "Sistema operativo", "Precio", "Ahorro"]
df = pd.DataFrame(portatiles, columns = columns)

df
```
